## Exploratory Data Analysis

### 1. Countries in order of total Covid Cases

```python
top_total_cases = df[['location', 'total_cases']].sort_values(by='total_cases', ascending=False).reset_index(drop=True) 
#arranging countries by descending order of total cases
```

```python
top_total_cases['location'].unique()
```

```python
sns.set_theme(style='white')
plt.figure(figsize=(10, 8))
# To avoid problem of different order after taking mean
order_list = ['United States', 'India', 'Brazil', 'United Kingdom', 'France',
       'Russia', 'Turkey', 'Italy', 'Mexico', 'South Africa', 'Canada', 'United Arab Emirates', 'Australia', 'Kenya', 'China']
#plotting
ax = sns.barplot(x='total_cases' , y='location', ci = None, data=top_total_cases, palette='rocket', order=order_list)
sns.despine(left=True, bottom=True)
ax.set(xticklabels=[]) 
ax.tick_params(bottom=False)
#labels
ax.set_title('Countries in order of average of total Covid cases from March-2020 to February-2022',fontsize=16, pad=15)
ax.set_xlabel('')
ax.set_ylabel('Countries',fontsize=14, labelpad=20)

#set annotation
for p in ax.patches:
    x_value = p.get_width()
    y_value = p.get_y() + p.get_height()/2
    ax.annotate(f'{int(x_value)}', xy=(x_value, y_value), 
                xytext=(10, 0), textcoords='offset points', ha='left', va='center', color='black', fontsize=11);
```

### 2. Analysing cases in top three countries

```python
# Converting to Date time
df['date'] = pd.to_datetime(df['date'])
```

```python
us_total = df[['date', 'total_cases', 'new_cases']].loc[df['location'] == 'United States']
india_total = df[['date', 'total_cases', 'new_cases']].loc[df['location'] == 'India']
brazil_total = df[['date', 'total_cases', 'new_cases']].loc[df['location'] == 'Brazil']
```

#### 2.1 Total Cases in US from 2020 March to 2022 February

```python
sns.set_context('notebook')
fig = plt.figure(figsize=(14, 6))
graph_2 = fig.add_subplot(111)
sns.set_theme(style='whitegrid')
graph_2 = sns.lineplot(x='date', y='total_cases', data=us_total, color = 'green', ls='--')

g2 = graph_2.twinx()
g2 = sns.lineplot(x='date', y='new_cases', data=us_total)

graph_2.set_ylabel('Total Cases (in tens of million)',fontsize=14, labelpad=20)
graph_2.set_title('Total Cases in United States from March-2020 to February 2022',fontsize=16, pad=15)
graph_2.set_xlabel('Date(March-2020 : February-2022)', fontsize=14, labelpad=15)
g2.set_ylabel('New Cases', fontsize=14, labelpad=20)

plt.show()
```

We can witness a slow rise in the beginning which picks up towards the end of 2020 and then begin to decrease towards mid of 2021 and then rises sharply towards end of 2021 and beginning of 2022 corresponding to the 'three covid waves'.

### 3. Comparing effect of  GDP Per Capita on  Total Deaths

```python
#creating a dataframe by grouping based on new deaths and gdp
death_and_gdp = pd.DataFrame({'Totaldeaths': df.groupby('location').sum()['new_deaths'], 'gdp': df.groupby('location')['gdp_per_capita'].mean()})
```

```python
death_and_gdp.reset_index(inplace=True)
```

```python
death_and_gdp
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(10, 8))
sns.lineplot(x='gdp', y='Totaldeaths', data=death_and_gdp)
plt.title('GDP vs Total Deaths', fontsize=16, pad=15)
plt.xlabel('GDP per capita', labelpad=15)
plt.ylabel('Total Deaths', labelpad=15)
plt.show()
```

- Most high cases of deaths are seen in the sub 30000 gdp countries. Only USA has high death rate in the above 50000 gdp category.
- Lack of proper data collection facilities in lower gdp countries might be lowering deaths in those cases.

```python
gdp_grp = df.groupby('location')['gdp_per_capita'].mean().sort_values(ascending=False).reset_index()
death_grp = df.groupby('location')['total_deaths'].mean().sort_values(ascending=False).reset_index()
```

### 6. Analysing the effect of vaccination drive in India

```python
india_covid = df.loc[df['location'] == 'India']
india_covid.head()
```

```python
india_covid.loc[india_covid['date'] > '2021-01-16', 'people_vaccinated'] = india_covid.loc[india_covid['date'] > '2021-01-16', 'people_vaccinated'].replace(to_replace=0, method='ffill')
```

```python
india_covid['percent_of_population_vaccinated'] = india_covid.loc[:, 'people_vaccinated'] / india_covid.loc[:, 'population']
```

```python
india_covid.iloc[-1].to_frame()[2:]
```

```python
india_covid[['date', 'population', 'people_vaccinated', 'percent_of_population_vaccinated']]
```

```python
after_1st_vaccine_date = india_covid[india_covid.date >'2021-01-16'] #data after first vaccination date
```

```python
lowest_new_cases = after_1st_vaccine_date.loc[after_1st_vaccine_date['new_cases'] > 1].sort_values(by = 'new_cases').head(1) #lowest cases after vaccination staretd
lowest_new_cases_date = lowest_new_cases.date # date of lowest cases
```

```python
lowest_new_cases['new_cases']
```

```python
lowest_new_cases_date
```

#### 6.1 Effect of Vaccination on New Cases

```python
sns.set_theme(style='ticks')
sns.set_context('notebook') #set scale of plot

fig = plt.figure(figsize=(14, 7))
ax = fig.add_subplot(111)
ax.plot(india_covid.date, india_covid.new_cases, 'darkblue', label = 'new cases', linewidth = 4, alpha= 0.6)

ax2 = ax.twinx() #create another y-axis that share the same x-axis
ax2.plot(india_covid.date, india_covid.percent_of_population_vaccinated, 'green', label = 'share of population vaccinated',linewidth = 5, alpha= 0.7)

ax.set_xlabel('date')
ax.set_ylabel(r'New Covid Cases', labelpad=20)
ax2.set_ylabel(r'Vaccinated Population as a Percentage', labelpad=20)

# vertical lines
plt.axvline(dt.datetime(2021,1,16), linewidth = 4, color='red',alpha= 0.3);
plt.axvline(lowest_new_cases_date, linewidth = 4, color='gold',alpha= 0.3);

ax.annotate('Lowest New Cases Date\n(2021-12-20)', xy=(lowest_new_cases_date, lowest_new_cases.new_cases),  xycoords='data',
             xytext=(-10, 80), textcoords='offset points',
             size=11, ha='right', va='top',
             bbox=dict(boxstyle='round', alpha=0.1),
             arrowprops=dict(arrowstyle='wedge,tail_width=0.5', alpha=0.1));

ax.annotate('First Vaccine \nDay in India\n(2021-01-16)', xy=(dt.datetime(2021,1,16), 15200),  xycoords='data',
             xytext=(9, 110), textcoords='offset points',
             size=11, ha='left', va='bottom',
             bbox=dict(boxstyle='round', alpha=0.1),
             arrowprops=dict(arrowstyle='wedge,tail_width=0.5', alpha=0.1))

ax.axes.set_title('How does the percentage of the population vaccinated \naffect new COVID-19 cases in India overtime?',fontsize=16, weight='bold', pad=40)


sns.despine(left=True, bottom=True)

fig.legend(loc=1, bbox_to_anchor=(0.32,1), bbox_transform=ax.transAxes);
```

- The first day of vaccination in India was on 2021-01-16.
-  We clearly see a decrease in new cases as the percentage of population receiving vaccination increases
- Even the `Omicron wave` that happened towards the end of 2021 and beginning of 2022 was not as bad as the second wave and quickly receded as seen in the graph as by this time, around 68% of the population received vaccinations.

#### 6.2Effect of Vaccination on New Deaths

```python
#New deaths time series
sns.set_theme(style='ticks')
sns.set_context('notebook')

fig = plt.figure(figsize=(14, 7))
ax = fig.add_subplot(111)
ax.plot(india_covid.date, india_covid.new_deaths, 'red', label = 'new deaths', linewidth = 4, alpha= 0.6)

ax2 = ax.twinx()
ax2.plot(india_covid.date, india_covid.percent_of_population_vaccinated, 'green', label = 'share of population vaccinated',linewidth = 4, alpha= 0.7)

ax.set_xlabel('date')
ax.set_ylabel(r'New Deaths', labelpad=20)
ax2.set_ylabel(r'Vaccinated Population as a Percentage', labelpad=20)

plt.axvline(dt.datetime(2021,1,16), linewidth = 4, color='blue',alpha= 0.3);
ax.annotate('First Vaccine \nDay in India\n(2021-01-16)', xy=(dt.datetime(2021,1,16), 181),  xycoords='data',
             xytext=(9, 80), textcoords='offset points',
             size=11, ha='left', va='bottom',
             bbox=dict(boxstyle='round', alpha=0.1),
             arrowprops=dict(arrowstyle='wedge,tail_width=0.5', alpha=0.1))

ax.axes.set_title('How does the percentage of the population vaccinated \naffect new deaths in India overtime?',fontsize=16, weight='bold', pad=40)

sns.despine(left=True, bottom=True)

fig.legend(loc=1, bbox_to_anchor=(0.32,1), bbox_transform=ax.transAxes);

```

**There is a significant decrease in new deaths after more than 40% of the population were vaccinated.**

### 8. Comparison of Population Age and Total Deaths

```python
age_grouped = df.groupby('location')[['median_age', 'aged_65_older', 'aged_70_older', 'total_deaths']].mean().reset_index()
```

```python
age_grouped
```

```python
for idx, val in enumerate(age_grouped['median_age']):
    if val < 30:
        age_grouped.loc[idx, 'median_age'] = 'young population'
    elif val < 40:
        age_grouped.loc[idx, 'median_age'] = 'middle aged population'
    else:
        age_grouped.loc[idx, 'median_age'] = 'older population'
```

```python
age_grouped = age_grouped.sort_values(by='total_deaths').reset_index()
```

```python
age_grouped
```

```python
fig = px.bar(
    age_grouped, x='total_deaths', y='location', color='median_age',
    labels = {'location': 'Countries', 'total_deaths': 'Death Count'}
)
fig.update_layout(barmode='stack', yaxis={'categoryorder':'total descending'},
                  title_text = 'Age of Population and Death Count',
                  title_font_size = 18,
                  title_font_color = 'Purple'
                  )
fig.show()
```

- More death counts seems to be in countries with a middle aged or young population
- This could be because majority of the population is working or out and about, and thus is more prone to contracting Covid.

### 10. Vaccination drive and extreme poverty

```python
poverty_vs_vaccine = df.groupby('location')[['extreme_poverty', 'people_fully_vaccinated']].mean().reset_index()
```

```python
poverty_vs_vaccine
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style='darkgrid')
sns.lineplot(x='extreme_poverty', y='people_fully_vaccinated', data=poverty_vs_vaccine)
plt.title('Percent of population in extreme poverty vs vaccination drive', pad=15, fontsize=17)
plt.xlabel('% of Population in Extreme Poverty', fontsize=14, labelpad=17)
plt.ylabel('People fully vaccinated', fontsize=14, labelpad=17)
plt.show()
```

- Higher number of people are fully vaccinated in countries with low extreme poverty.
- We see a downward trend in number of people fully vaccinated as percent of population living in extreme poverty increases
- Only India with more than half of the population fully vaccinated is an outlier here; perhaps due to the focus and attention the vaccination drive received in the country.
